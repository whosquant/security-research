###  持有期收益率计算

```python
import pandas as pd
```

```python
dfprice = pd.read_excel('企业债价格.xlsx',header = 1)
dfprice.Date = dfprice.Date.astype(str).apply(lambda x:x[:-3])
dfprice.set_index('Date',inplace = True)
```

```python
df_interestpay = pd.read_excel('企业债付息日和付息率2.xlsx')
df_interestpay.columns = ['证券代码', '证券简称', '利率说明', '付息日','起息日']
df_interestpay.set_index('证券代码',inplace = True)
```

```python
df_net_price = pd.read_excel('债券净值.xlsx',header = 1)
df_net_price.Date = df_net_price.Date.astype(str).apply(lambda x:x[:-3])
df_net_price.set_index('Date',inplace = True)
```

### 制作利率的模板

```python
df_interestpay['len'] = df_interestpay['利率说明'].apply(lambda x:len(x))#.hist()
bondlist = df_interestpay[(df_interestpay.len<=20)|(df_interestpay.len>=30)].index
df_interestpay = df_interestpay.loc[df_interestpay.index.isin(bondlist)]
df_interestpay[df_interestpay.index =='139415.SH'].利率说明.values
```

```python
list1 = []
for i in range(2010,2022):
    for k in range(1,13):
        str1 = str(i)+'-'+str(k).zfill(2)
        list1.append(str1)
df0 = pd.DataFrame(index = list1)
```

```python
def get_tax_rate_series(code = '1480527.IB'):
    df1 = df0.copy()
    df1[code] = 0
    text = df_interestpay.loc[df_interestpay.index == code]['利率说明'].values[0]
    paymonth = df_interestpay['付息日'][code]
    paymonth = int(paymonth[:paymonth.find('月')])
    qxyear = str(df_interestpay['起息日'][code])[:4]
    if len(text) <= 10:
        taxrate = float(text[:-1])
        for i in df1.index.tolist():
            if int(i[-2:]) ==  paymonth:
                if int(i[:4]) > int(qxyear):
                    df1[code][i] = taxrate
#        return df1
    elif len(text) >= 30:
        list1 = text.split(';')
#         print(list1)
        #listdate = [(i[:4]+'-'+i[4:6]+'-'+i[6:8]) for i in list1]
        startmonthlist = [(i[:4]) for i in list1]
        endmonthlist = [(i[9:13]) for i in list1]
#         print(startmonthlist)
#         print(endmonthlist)
        list2 = [i.find(':')+1 for i in list1]
        list3 = [i.find('%') for i in list1]
        dict1 = {}
        for i in range(len(list1)):
            dict1[startmonthlist[i]] = float(list1[i][list2[i]:list3[i]])
#         print(dict1)
        for i in df1.index.tolist():
            year = int(i[:4])
            month = int(i[-2:])
#             print(1)
            if int(i[-2:]) ==  paymonth:
                if year<int(startmonthlist[0]):
                    taxrate = 0
                else:
                    for j in range(len(list1)):

                        if int(startmonthlist[j])< year and int(endmonthlist[j] )>=year:
                            taxrate = dict1[startmonthlist[j]]
#                             print(taxrate)
                df1[code][i] = taxrate
    else:
        print('!!!')
    return df1[code]
a = get_tax_rate_series()
#a.to_excel('demo2.xlsx')
```

```python
df_total_tax = df0.copy()
for i in bondlist:
#     print(i)
    s1 = get_tax_rate_series(i)
    df_total_tax =pd.concat([df_total_tax,s1],axis = 1)
```

```python
dfinterest = (df_net_price.shift(1)*df_total_tax)/100
```

#### 债券复权价格 = (初始净值 - 现在净值)+现在全价 +利息累计支出

```python
(-df_net_price+100+dfprice+dfinterest.cumsum()).to_excel('FINALPRICE.)
```
